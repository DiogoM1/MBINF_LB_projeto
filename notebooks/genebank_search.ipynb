# Verificar e analisar a informação complementar fornecida pela lista de features e seus qualifiers

```python
print(seq_r.name)
for feature in seq_r.features:
    print('feature:', feature.type)
    for Key,Value in feature.qualifiers.items():
        print(Key,": ", Value)
    print("")
```
